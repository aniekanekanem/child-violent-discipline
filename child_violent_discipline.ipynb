```python
import pandas as pd
import numpy as np
```

```python
violent_discipline_data = pd.read_excel("XLS_Violent-discipline-database_June-2023.xlsx", header=9)
```

```python
# violent_discipline_data = violent_discipline_data.drop(index=0)
```

```python
violent_discipline_data.head(15)
```

```python
violent_discipline_data = violent_discipline_data[['Unnamed: 0','Unnamed: 1','Unnamed: 3','Unnamed: 5']]
violent_discipline_data
```

```python
violent_discipline_data.rename(columns={
    'Unnamed: 0':'Country',
    'Unnamed: 1':'Any violent discipline',
    'Unnamed: 3':'Male',
    'Unnamed: 5':'Female'
},inplace=True)
```

```python
violent_discipline_data.head()
```

```python
violent_discipline_data.replace('-', np.nan, inplace=True)
violent_discipline_data
```

```python
violent_discipline_data = violent_discipline_data.dropna()
violent_discipline_data
```

```python
violent_discipline_data = violent_discipline_data.reset_index(drop=True)
violent_discipline_data
```

```python
violent_discipline_data.describe()
```

```python
violent_discipline_data['Any violent discipline'] = pd.to_numeric(violent_discipline_data['Any violent discipline'], errors='coerce')

violent_discipline_data['Any violent discipline'].apply(pd.to_numeric, errors='coerce').notnull().all()
violent_discipline_data.describe()
```

```python
# Declaring the first 20 dataset
first_20 = violent_discipline_data[0:20]
first_20.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the Seaborn style
sns.set(style="whitegrid") # white, dark, whitegrid, darkgrid, ticks

# Create a bar plot
plt.figure(figsize=(10,6))
ax = sns.barplot(x='Any violent discipline', y='Country', data=first_20)

# Add title and labels
plt.title('Country V violent discpline rate [Part1]')
plt.xlabel('Any violent discipline')
plt.ylabel('Country')

# Displaying population values on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_width():,.0f}',  # Display the rate with commas for thousands
                (p.get_width(), p.get_y() + p.get_height() / 2.),  # Position at the end of the bar
                ha='left', va='center',  # Align the text to the left and vertically centered
                fontsize=10, color='black',  # Text properties
                xytext=(5, 0),  # Slightly offset text to avoid overlap with the bar
                textcoords='offset points')

# Show plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
